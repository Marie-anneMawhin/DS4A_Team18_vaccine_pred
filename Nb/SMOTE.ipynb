## DS4A Project - Team 18 - Vaccine Acceptance_SMOTE

```python
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# 1. Loading data, merging df

```python
labels = pd.read_csv('Data/training_set_labels.csv')
imp_feat = pd.read_csv('Data/imputed_train_hot_encoded.csv')
labels.head()
```

```python
imp_feat.head()
```

```python
labels.set_index('respondent_id', inplace=True)
imp_feat.set_index('Unnamed: 0', inplace=True)
imp_feat.sort_index(inplace=True)
```

```python
merged_df = imp_feat.join(labels)
```

```python
df_h1n1 = merged_df.reset_index(drop=True).drop(['seasonal_vaccine'], axis=1)
df_h1n1.head(20)
```

# 2. Balancing classes for h1n1_vaccine using SMOTE (over-sampling for minority class) 

```python
from imblearn.over_sampling import SMOTE
```

```python
X = np.asarray(df_h1n1.iloc[:, :-1])
y = np.asarray(df_h1n1["h1n1_vaccine"])
print(X.shape)
print(y.shape)
```

```python
print(X[9], y[9])
```

```python
print("Before Oversampling, counts of label '1': {}".format(sum(y==1)))
print("Before Oversampling, counts of label '0': {} \n".format(sum(y==0)))

sm = SMOTE(random_state=13)
X_res, y_res = sm.fit_resample(X, y)

print('After Oversampling, shape of X: {}'.format(X_res.shape))
print('After Oversampling, shape of y: {} \n'.format(y_res.shape))

print("After Oversampling, counts of label '1': {}".format(sum(y_res==1)))
print("After Oversampling, counts of label '0': {}".format(sum(y_res==0)))
```

# 3. Converting balanced data into pd DataFrame 


```python
df_X_res = pd.DataFrame(data=X_res, columns=df_h1n1.columns [:-1])
df_X_res.head(20)
```

```python
df_X_res.shape
```

```python
df_y_res = pd.DataFrame(data=y_res, columns=['h1n1_vaccine'])
```

```python
df_y_res.head(20)
```

```python
df_y_res.shape
```

```python
imp_feat_h1n1_bal = df_X_res.join(df_y_res)
```
